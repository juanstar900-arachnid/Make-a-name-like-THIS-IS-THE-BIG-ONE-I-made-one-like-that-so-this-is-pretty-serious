# PART 2: DEUTSCH-JOZSA ALGORITHM

## The Problem It Solves

**What is the mystery box problem?**

You are given a black-box function that takes an input and returns either 0 or 1. The function is guaranteed to be either constant or balanced, and the goal is to figure out which one it is

**Constant function means:**

The function outputs the same value (always 0 or always 1) for every possible input

**Balanced function means:**

The function outputs 0 for half of the inputs and 1 for the other half

---

## Classical vs Quantum Solution

### Classical Computer Approach:

**Why can't it do better?**

Because without checking more than half the inputs, a classical computer canâ€™t rule out the possibility that the function changes later

### Quantum Computer Approach:

**How many checks does a quantum computer need?**

One check.

**How is this possible?**

The quantum computer evaluates all inputs at once using superposition, so, one.
