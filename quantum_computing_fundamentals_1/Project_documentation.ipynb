### [ ] Entanglement  
**How it applies to my project:**
[Explain how qubits that can't be described independently help solve your problem]

 Because the qubits cannot be described independently, any attempt by a hacker to "intercept" or "measure" one qubit instantly collapses the entanglement and changes the state of the other. This creates a physical "alarm" that alerts the system to the breach before any data is actually stolen.
---

### [ ] Interference

Interference is the idea that quantum states behave like waves. When a qubit can be in a position of 0 and 1, each possible state acts like a wave. These waves interfere with each other. Constructive interference waves reinforce each other, and destructive interference waves cancel each other out. In cyber security this process can be used to destruct "bad paths" and let "good paths" survive, furthermore, it's what makes algorithms like Shor's algorithm and the Deutsch-Jozsa algorithm work, which lets us simulate attackers.


**Example:** Interference amplifies molecular structures with desired properties while canceling out poor candidates, reducing the search space dramatically.

### [ ] Deutsch-Jozsa Algorithm
**How it applies to my project:**
[Explain how this algorithm concept connects - even if just as a demonstration of quantum speedup]
The Deutsch-Jozsa algorithm shows how a quantum computer can determine a global property of a function in one step instead of many, demonstrating the type of speedup that could be useful for analyzing security data and detecting patterns.
